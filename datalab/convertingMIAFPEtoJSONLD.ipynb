# Template for conversion
We need to define a data model for the conversion. There we define the linked structure via the rdf paradigm (subject-predicate-object). For example:

ppeo:Investigation -> ppeo:hasStudy -> ppeo:Study
ppeo:Study -> ppeo:hasObservationUnit -> ppeo:ObservationUnit
ppeo:ObservationUnit -> ppeo:isPartOf -> SOSA:Observation
SOSA:Observation -> SOSA:isObservedBy -> SOSA:Sensor
SOSA:Sensor -> SOSA:isHostedBy -> SOSA:Platform
SOSA:Sensor -> SOSA:isPartOf -> SOSA:System

We therefore need to define the data model and then depending on the values given within a MIAPPE file construct a JSON-LD file.

## How could this be implemented?

- Each sheet within the MIAPPE file is read for each row and converted into a very basic JSON object just containing an ID, a type (and correponding to the type a description and a set of synonyms).

- Identifiers are then used to link the different objects. The link is defined by a property, which is defined by the data model.
