## Sensors and Platforms

HyPlant imaging spectrometer consists of two sensor-modules:

- The DUAL module records hyperspectral data in the spectral range from 400-2500nm, which is useful in deriving biochemical and structural plant properties.

- The FLUO module acquires data in the red and near-infrared bands in the spectral range of 670-780nm with a distinctly higher spectral resolution and sampling interval. 

Hyperspectral imaging has the benefit of cost-efficient and non-destructive assessment of structural (e.g. leaf area index, biomass) or biochemical (e.g. pigment or water content) vegetation properties. 




**FLOX Systems:**
FLOX systems are used to measure top-of-canopy reflectance and solar-induced fluorescence (SIF) using high-performance spectrometers

The FLOX was developed by the IBG-2: Plant Sciences of the JÃ¼lich Research Center and the Remote Sensing of Environmental Dynamics Laboratory of the University Milano-Bicocca. In 2015, two members of the partners founded the new start-up company JB Hyperspectral http://jb-hyperspectral.com/wordpress/. The commercial versions of the FLOX Boxes are distributed by this young start-up company.
FLOX is measuring the sun-induced chlorophyll fluorescence (SIF) of plant canopies as a measure of plant photosynthetic capacity under natural light conditions. A QEPro spectrometer from Ocean Optics covering the Red/Near Infrared region (650 â 800 nm) measures chlorophyll fluorescence of plant canopies in the wavelength region of the two atmospheric oxygen absorption bands, O2B and O2A (689 nm and 760 nm, respectively), with a high spectral resolution. SIF of plant canopies can be constantly monitored by the comparison of sequential measurements of light reflected and emitted from the plant canopy using a downward fiberoptics and solar irradiance using an upward directed reference fiberoptics. The spectrometers are embedded in a temperature controlled housing to maintain a stable level of dark current of the spectrometers. An accurate automatic optimization of the measurement duration of both channels to prevailing light conditions is applied to maximize the signal-to-noise ratio and minimize acquisition time. Thereby, FLOX can be used to constantly monitor SIF under changing environmental conditions.

The Shoot group at IBG-2 currently operates 5 FLox systems, some of which are used as stationary systems to continuously record solar-induced fluorescence (SIF) and hyperspectral reflectance from selected study sites and which thus produce time-series of vegetation functioning over the season. Other FLOX systems ae integrated in sensor positioning platforms and are operated within various projects to measure SIF and vegetation reflectance across experimental field systems. One FLOX system ist currently being modified in cooperatin with the manufacturer to be mounted on a UAV platform to enable SIF measurements from low flying plattforms.

Additionally, we are operating 2 ROX systems, which are smaller versions consisting of only one spectrometer. The ROX systems measure canopy reflectance but cannot be used to retrieve solar-induced fluorescence. The ROX systems are used as small weight field spectrometers to record canopy and leaf reflectance.

FLOX and ROX can be combined with the leaf-clip FluoWAT, which is a custom made leaf clip to measure leaf-reflecance and transmittance as well as upwelling and down-welling solar induced fluorescence from single leaves in the field.





```python
# Gatherin information from DataPub

import os
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime


class Dataset_DataPub:
    UNIT_MULTIPLIERS = {
        'B': 1,
        'K': 1024,
        'M': 1024**2,
        'G': 1024**3,
        'T': 1024**4,
        'P': 1024**5
    }

    def __init__(self, root_url, auto_exploration = True):
        """
        General repository handling class for scraping and data collection.
        
        Args:
            root_url (str): Root URL of the repository to scrape.
            auto_exploration (bool): if True autmatically scrape for data
        """
        self.root_url = root_url
        self.file_data = []
        if auto_exploration:
            self.explore_folders()

    def _fetch_html(self, url):
        """Fetch HTML content from a URL."""
        try:
            response = requests.get(url)
            response.raise_for_status()
            return BeautifulSoup(response.text, 'html.parser')
        except requests.RequestException as e:
            print(f"Error fetching URL {url}: {e}")
            return None

    def _parse_row(self, row, base_url):
        """
        Extracts data from a table row and adds it to file_data.

        Args:
            row (Tag): A BeautifulSoup tag representing a table row.
            base_url (str): The base URL of the current directory.
        """
        columns = row.find_all('td')[1:]
        if len(columns) != 4:
            return None

        href = columns[0].find("a").get("href")
        item_type = "folder" if href.endswith("/") else "file"

        if item_type == "file":
            size_match = re.match(r'([0-9]*\.?[0-9]+)([KMGTP]?)', columns[2].text.strip())
            size, unit = (float(size_match.group(1)), size_match.group(2) or 'B') if size_match else (0, 'B')

            # Convert size to bytes
            size_in_bytes = size * self.UNIT_MULTIPLIERS.get(unit.upper(), 1)

            # Extract file format, file name, and folder
            file_url = os.path.join(base_url, href)
            file_name = href.split("/")[-1]
            file_format = file_name.split(".")[-1] if "." in file_name else None
            folder = "/".join(file_url.split("/")[:-1])

            # Add extracted information to file_data
            self.file_data.append({
                "url": file_url,
                "last_mod": columns[1].text.strip(),
                "size": size,
                "unit": unit,
                "size_in_bytes": size_in_bytes,
                "desc": columns[3].text.strip() if columns[3].text.strip() != "\xa0" else None,
                "file_name": file_name,
                "file_format": file_format,
                 "folder" : "/".join(file_url.split("/")[len(self.root_url.split("/"))-1: -1])  # Relative folder path
            })

        return href if item_type == "folder" else None

    def explore_folders(self, current_url=None):
        """
        Recursively explore folders and collect file data.

        Args:
            current_url (str): The current URL being explored. Defaults to root_url.
        """
        if current_url is None:
            current_url = self.root_url

        soup = self._fetch_html(current_url)
        if not soup:
            return

        rows = soup.find("table").find_all("tr")[3:]  # Skip header rows
        for row in rows:
            href = self._parse_row(row, current_url)
            if href:  # If it's a folder, recurse
                self.explore_folders(os.path.join(current_url, href))

    def to_dataframe(self):
        """Convert collected file data to a pandas DataFrame."""
        if not self.file_data:
            print("No file data to convert.")
            return pd.DataFrame()
        return pd.DataFrame(self.file_data)

    def save_to_csv(self, name, path = "data"):
        """
        Save the data to a CSV file with a name generated based on the current date and time.

        Args:
            path (str): The directory to save the CSV file.
            name (str): The base name of the file.
        """
        if not os.path.exists(path):
            os.mkdir(path)
        df = self.to_dataframe()
        if df.empty:
            print("No data to save.")
            return

        # Generate the timestamp
        timestamp = datetime.now().strftime("%Y%m%d%H%M")

        # Create the full file name
        csv_file = os.path.join(path, f"{name}_{timestamp}.csv")

        # Save the DataFrame to a CSV file
        df.to_csv(csv_file, index=False)
        print(f"Data saved to {csv_file}")





# url = 'https://datapub.fz-juelich.de/plantsciences/HyData-HyPlant-FLUO/'


# dataset = Dataset_DataPub(url)
# dataset.save_to_csv("HyPlant")
```

```python
import pandas as pd
import json

class DictPath(dict):
    """
    A dictionary subclass with extended functionality for nested operations.
    """

    def get_nested_value(self, path, separator="/"):
        """
        Retrieve a value from the dictionary using a path-like string.

        Args:
            path (str): The path to the value, e.g., "a/1".
            separator (str): The separator used in the path, default is "/".

        Returns:
            any: The value found at the path, or None if the path is invalid.
        """
        keys = path.split(separator)
        current = self

        try:
            for key in keys:
                current = current[key]
            return current
        except (KeyError, TypeError):
            return None

    def fetch_all_values(self, parent_path="", separator="/"):
        """
        Recursively fetch all values in the dictionary and return their paths and values.

        Args:
            parent_path (str): The current path in the recursion.
            separator (str): The separator used for path construction.

        Returns:
            list: A list of tuples containing the path and the value.
        """
        result = []

        for key, value in self.items():
            current_path = f"{parent_path}{separator}{key}" if parent_path else key

            if isinstance(value, dict):
                # Recurse into the dictionary
                result.extend(DictPath(value).fetch_all_values(current_path, separator))
            else:
                result.append((current_path, value))

        return result

    def write_nested_value(self, path, value, separator="/", append = True):
        """
        Write a value into the dictionary at the specified path.

        Args:
            path (str): The path to the value, e.g., "a/1".
            value (any): The value to set at the specified path.
            separator (str): The separator used in the path, default is "/".
            append : adds a value to a list at the desired path in the structure

        Returns:
            DictPath: The modified dictionary.
        """
        keys = path.split(separator)
        current = self

        for key in keys[:-1]:
            if key not in current or not isinstance(current[key], dict):
                current[key] = DictPath()  # Create nested ExtendedDicts as needed
            current = current[key]
        if append:
            if isinstance(current[keys[-1]], list):
                current[keys[-1]].append(value)
            else:
                current[keys[-1]] = []
        else:   
            current[keys[-1]] = value
        return self

class ARCManager:
    """
    A class to generate and manage an ARC structure from a list of data files (CSV).
    """

    def __init__(self, csv: str, data_structure=None):
        self.data = pd.read_csv(csv)
        self.data_structure = DictPath(data_structure or {})
        self.label_rules = []

    def add_data_structure(self, data_structure: dict):
        """
        Dynamically add or update a custom ARC data structure.

        Args:
            data_structure (dict): The data structure to add or update.
        """
        if not isinstance(data_structure, dict):
            raise ValueError("data_structure must be a dictionary.")
        self.data_structure = DictPath(data_structure)


    def import_arc_structure_txt(self, txtfile: str):
        """
        Import an ARC structure from a text file.

        Args:
            txtfile (str): Path to the text file containing ARC structure paths.

        Example file format:
            studys/study0001
            studys/study0002
            assays/subcategory01/assay01
        """
        if not txtfile.endswith(".txt"):
            raise ValueError("Input file must be a .txt file.")
        
        self.data_structure = DictPath()
        with open(txtfile, "r") as file:
            for line in file.readlines():
                self.data_structure.write_nested_value(line.strip(), [])
        print(f"Imported ARC structure from {txtfile}")


    def add_label_rule(self, func):
        """Adding a rule onto the dataframe, which is adding the arc_ammping entry"""
        self.label_rules.append(func)


    def apply_label_rules(self):
        for rule in self.label_rules:
            self.data = rule(self.data)


    def save_dataframe_to_csv(self, name, path="data"):
        """
        Save the data to a CSV file with a name generated based on the current date and time.

        Args:
            path (str): The directory to save the CSV file.
            name (str): The base name of the file.
        """
        if not os.path.exists(path):
            os.mkdir(path)


        # Generate the timestamp
        timestamp = datetime.now().strftime("%Y%m%d%H%M")

        # Create the full file name
        csv_file = os.path.join(path, f"{name}_{timestamp}_ARCLabeled.csv")

        # Save the DataFrame to a CSV file
        self.data.to_csv(csv_file, index=False)
        print(f"Data saved to {csv_file}")

    def fill_data_structure(self):
        """For this method the labeling needs to be done"""
        for _, row in self.data.iterrows(): 
            try:
                self.data_structure.write_nested_value(row["arc_mapping"], row["url"])
            except:
                print(row["arc_mapping"])
    

    def create_bundles(self):
        pass

    def save_data_structure_to_json(self, filepath):
        """
        Save the current data structure (arc_structure) to a JSON file.

        Args:
            filepath (str): The file path where the JSON will be saved.
        """
        try:
            with open(filepath, 'w', encoding='utf-8') as json_file:
                json.dump(self.data_structure, json_file, ensure_ascii=False, indent=4)
            print(f"Data structure successfully saved to {filepath}")
        except Exception as e:
            print(f"Error saving data structure to JSON: {e}")

    def create_folder_structure(self, path = "../ARC", from_datastructure = True):
        """Helps in generating a folder structure for your ARC"""
        if from_datastructure:
            for folder, file in self.data_structure.fetch_all_values():
                if not os.path.exists(os.path.join(path, folder)):
                    os.makedirs(os.path.join(path, folder))
        else:
            for folder in self.data["arc_mapping"].unique():
                if not os.path.exists(os.path.join(path, folder)):
                    os.makedirs(os.path.join(path, folder))



    
# Creating an ARC on given structure and rules

arc_structure = {
    "meta" :[],
    "studys" : {
        "2018_SEL" : [],
        "2019_WST" : [],
        "2019_NRS" : [],
        "2019_SEL" : [],
        "2019_TR32" : [],
        "2020_CKA" : [],
        "2021_CKA" : [],
        "2021_HOE" : [],
        "2021_GLO" : [],
        "2022_HOE" : [],
        "2022_CKA" : [],
        "2022_INN" : [],
        "2022_SEL" : [],
        "2023_JIM" : []
    },
    "assays" : {
        "meta" : [],
        "FLUO" : {
            "2018_SEL_0600" : [],
            "2019_WST_1500" : [],
            "2019_NRS_1500" : [],
            "2019_SEL_0600" : [],
            "2019_TR32_1800" : [],
            "2020_CKA_0350" : [],
            "2020_CKA_0600" : [],
            "2021_CKA_0350" : [],
            "2021_HOE_0600" : [],
            "2021_GLO_1150" : [],
            "2022_HOE_0600" : [],
            "2022_CKA_0350" : [],
            "2022_INN_0600" : [],
            "2022_SEL_0600" : [],
            "2023_JIM_0600" : []
        },
        "DEMs" : {
            "2018_SEL_0600" : [],
            "2019_WST_1500" : [],
            "2019_NRS_1500" : [],
            "2019_SEL_0600" : [],
            "2019_TR32_1800" : [],
            "2020_CKA_0350" : [],
            "2020_CKA_0600" : [],
            "2021_CKA_0350" : [],
            "2021_HOE_0600" : [],
            "2021_GLO_1150" : [],
            "2022_HOE_0600" : [],
            "2022_CKA_0350" : [],
            "2022_INN_0600" : [],
            "2022_SEL_0600" : [],
            "2023_JIM_0600" : []
        },
        "FLOX" : {
            "2018_SEL" : [],
            "2019_WST_NRS" : [],
            "2020_CKA" : [],
            "2021_CKA" : [],
            "2021_GLO" : [],
            "2022_CKA" : [],
        }
    },
    "workflows" : {
        "FLUO" : [],
        "FLOX" : [],
        "DEMS" : [],
        "SFM" : [],
        "SFMNN" : []
    },
    "runs" : {
        "meta" : [],
        "SFM" : {
            "2018_SEL_0600" : [],
            "2019_WST_1500" : [],
            "2019_TR32_1800" : [],
            "2020_CKA_0350" : [],
            "2020_CKA_0600" : [],
            "2021_CKA_0350" : [],
            "2022_CKA_0350" : [],
            "2022_CKA_0600" : [],
            "2022_INN_0600" : []
        },
        "SFMNN" : {
            "2018_SEL_0600" : [],
            "2019_WST_1500" : [],
            "2019_NRS_1500" : [],
            "2019_TR32_1800" : [],
            "2020_CKA_0350" : [],
            "2020_CKA_0600" : [],
            "2021_CKA_0350" : [],
            "2021_GLO_1150" : [],
            "2022_CKA_0350" : [],
            "2022_INN_0600" : [],
            "2023_JIM_0600" : []
        }
    }
}

def detect_label(dataframe :pd.DataFrame):
    """
    Acts on a DataFrame with the following features:
        - url: Full path of the data file.
        - file_format: File type/format.
        - name: File name.

    The function adds a column 'arc_mapping' with the ARC-compliant label for each row.

    Args:
        dataframe (pd.DataFrame): Input DataFrame containing the file information.

    Returns:
        pd.DataFrame: DataFrame with an additional 'arc_mapping' column.
    """

    def detect_label_inner(path, fformat, ffname):
        """Detects ARC labels based on the rules provided."""
        # Metadata detection
        if not isinstance(path, str):
            return "meta"
        if "01_data" not in path and "02_products" not in path:
            return "meta"

        # Assays detection
        if "01_data" in path:
            if fformat in ["txt", "md"] and "README" in ffname:
                return "assays/meta"
            if "radiance" in path:
                return "assays/FLUO"
            if "FLOX" in path:
                return "assays/FLOX"
            if "DEMs" in path:
                return "assays/DEMs"

        # Runs detection
        if "02_products" in path:
            if fformat in ["txt", "md"] and "README" in ffname:
                return "runs/meta"
            if "SFMNN" in path:
                return "runs/SFMNN"
            if "SFM" in path:
                return "runs/SFM"

        return None

    # def extract_year_location_altitude(filename):
    #     """Extract year, location, and altitude from the file name."""
    #     # match = re.match(r"(\d{4})\d{4}-([A-Z0-9]{3,4})-\d{4}-(\d{3,4})", filename)
    #     match = re.search(r"(\d{4})(\d{2})(\d{2})-([A-Z]{3,4})-\d{4}-(\d{4})", filename)
    #     if match:
    #         year, location, altitude = match.groups()
    #         year, month, day, location, altitude = match.groups()
    #         altitude = altitude.zfill(4)  # Normalize altitude to 4 digits
    #         return f"{year}_{location}_{altitude}"
        # return None

    def extract_year_location_altitude(filename):
        """
        Extract year, location, and altitude from the file name.

        Args:
            filename (str): The file name or path.

        Returns:
            str: A formatted string containing year, location, and altitude,
                or None if no match is found.
        """
        # Match filenames with a pattern like '20190627-TR32-1308-1800'
        match = re.match(r"(\d{4})\d{4}-([A-Z0-9]{3,4})-\d{4}-(\d{3,4})", filename)
        if match:
            year,  location, altitude = match.groups()
            altitude = altitude.zfill(4)  # Normalize altitude to 4 digits
            # Apply specific rules for known anomalies
            return f"{year}_{location}_{altitude}"
        else:
            print(filename)
        return None

    def sort_data(path, fformat, fname):
        """Sort data into ARC-compliant labels."""
        labels = detect_label_inner(path, fformat, fname)

        if not labels:
            return "None"

        sublabel = extract_year_location_altitude(fname)
        if "meta" in labels:
            return labels
        if sublabel:
            return f"{labels}/{sublabel}"
        elif labels == "assays/FLOX":
            if "21-" in path:
                return f"{labels}/2018_SEL"
            elif "22-" in path:
                return f"{labels}/2019_WST_NRS"
            elif "23-" in path:
                return f"{labels}/2020_CKA"
            elif "25-" in path:
                return f"{labels}/2021_CKA"
            elif "26-" in path:
                return f"{labels}/2021_GLO"
            elif "28-" in path:
                return f"{labels}/2022_CKA"
            else:
                return f"assays/meta"
        else:
            return f"meta"

    # Apply the sorting logic to the DataFrame and store results in the 'arc_mapping' column
    dataframe["arc_mapping"] = dataframe.apply(
        lambda row: sort_data(row["folder"], row["file_format"], row["file_name"]), axis=1
    )

    return dataframe
   
    



new_arc = ARCManager(r"C:\Users\l.grygosch\Desktop\FAIRAgro\developement\MIAFPE\benchmark\HyPlant\Scripts\data\arc\HyPlant_202412041752_ARCLabeled.csv")

# Adding Structure
new_arc.add_data_structure(arc_structure)

# new_arc.add_label_rule(detect_label)

# new_arc.apply_label_rules()

new_arc.create_folder_structure()

# new_arc.save_dataframe_to_csv("HyPlant", "data/arc")

# new_arc.fill_data_structure()


# new_arc.save_data_structure_to_json("HyPlant.json")

```
